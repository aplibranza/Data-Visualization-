#Case Study 1: Netflix and Subscriber Growth
Netflix is often used as an example of a data-driven company. In the notebook, subscriber data is plotted to show how the platform grew across regions.

```python
import matplotlib.pyplot as plt

regions = ["North America", "Europe", "Latin America", "Asia-Pacific"]
subscribers = [75, 68, 39, 25]

plt.figure(figsize=(8,6))
bars = plt.bar(regions, subscribers, color=['red','blue','green','purple'])
plt.title("Netflix Subscribers by Region (millions)")
plt.ylabel("Subscribers (millions)")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height()-5,
             str(bar.get_height()),
             ha='center', color='white', fontsize=12)

plt.show()
```

# Case Study 2: Spotify and Music Preferences
Another example looks at Spotify, where user behavior is mapped across genres. Instead of a pie chart, let's visualize Spotify streams as a stacked bar chart to highlight percentage share.

```python
import matplotlib.pyplot as plt

# Data
genres = ["Pop", "Hip-Hop", "Rock", "EDM", "Classical"]
streams = [120, 95, 80, 60, 25]
total = sum(streams)
percentages = [s / total * 100 for s in streams]

# Colors
colors = ['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd']

# Plot
fig, ax = plt.subplots(figsize=(10, 2.5))
left = 0
for genre, pct, color in zip(genres, percentages, colors):
    ax.barh(0, pct, left=left, color=color, edgecolor="white")
    ax.text(left + pct/2, 0, f"{genre}\n{pct:.1f}%",
            ha='center', va='center', color="white", fontsize=9, weight="bold")
    left += pct

# Formatting
ax.set_xlim(0, 100)
ax.set_ylim(-0.6, 0.6)
ax.set_xlabel("Percentage of Total Streams")
ax.set_yticks([])  # Hide y-axis
ax.set_xticks(range(0, 101, 10))  # Show ticks every 10%
ax.set_title("Spotify Streams by Genre (Share of Total)", fontsize=14, weight="bold")

plt.show()
```

# Case Study 3: Amazon Transactions Across Regions
Amazon's case shows how e-commerce thrives in different parts of the world.

```python
regions = ["US", "Europe", "Asia", "Latin America"]
transactions = [1200, 850, 940, 400]

plt.figure(figsize=(8,6))
bars = plt.bar(regions, transactions, color=['cadetblue','orange','green','purple'])
plt.title("Amazon Transactions per Region")
plt.ylabel("Transactions")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height()+20,
             str(bar.get_height()),
             ha='center')

plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Data
regions = ["US", "Europe", "Asia", "Latin America"]
transactions = [1200, 850, 940, 400]

# Map regions to representative countries for visualization
mapping = {
    "US": "United States",
    "Europe": "Germany",       # proxy for Europe
    "Asia": "India",           # proxy for Asia
    "Latin America": "Brazil"  # proxy for Latin America
}

df = pd.DataFrame({
    "Region": regions,
    "Country": [mapping[r] for r in regions],
    "Transactions": transactions
})

# Custom colors
custom_colors = {
    "US": "cadetblue",
    "Europe": "orange",
    "Asia": "green",
    "Latin America": "purple"
}

# Choropleth map with custom colors
fig = px.choropleth(
    df,
    locations="Country",
    locationmode="country names",
    color="Region",  # Use Region for discrete coloring
    hover_name="Region",
    hover_data={"Transactions": True, "Country": False},
    projection="natural earth",
    title="Amazon Transactions Across Regions",
    color_discrete_map=custom_colors
)

fig.show()
```

# Technical Breakdown: Building Stories with Python
The notebook shows how simple Python libraries: Matplotlib and Seaborn are powerful tools for visual storytelling.

A heatmap example (for a delivery company):

```python
import seaborn as sns
import numpy as np

data = np.array([[30, 55, 80],
                 [45, 65, 90],
                 [20, 40, 70]])

sns.heatmap(data, annot=True, cmap="YlOrRd", xticklabels=["Morning","Afternoon","Evening"], yticklabels=["Zone A","Zone B","Zone C"])
plt.title("Delivery Density by Zone and Time of Day")
plt.show()
```
